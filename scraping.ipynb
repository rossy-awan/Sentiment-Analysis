```python
%%capture
!pip install google_play_scraper
from google.colab import drive
from google_play_scraper import Sort, reviews_all
import pandas as pd
```

Scraping

```python
reviews_df = pd.DataFrame(reviews_all('org.brilliant.android', lang='en', country='us', sort=Sort.MOST_RELEVANT, filter_score_with=None))
print(reviews_df.shape)
reviews_df.head(3)
```

Cleaning

```python
reviews_df = reviews_df.drop(columns=['reviewId', 'userName', 'userImage', 'thumbsUpCount', 'reviewCreatedVersion', 'at', 'replyContent', 'repliedAt', 'appVersion'])
reviews_df = reviews_df.dropna()
reviews_df.drop_duplicates(subset='content', inplace=True)
```

Export Dataset

```python
drive.mount('/content/drive')
reviews_df.to_csv('/content/drive/My Drive/Colab Notebooks/dataset-brilliant.csv', index=None, header=True)
```
